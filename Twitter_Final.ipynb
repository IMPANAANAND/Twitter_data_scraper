```python
from ntscraper import Nitter
import pandas as pd

# Initialize the scraper
scraper = Nitter(0)

# List of usernames
usernames = ['narendramodi', 'elonmusk', 'klrahul']

# Prepare an empty dictionary to collect data
tweets_dict = {}
profiles_dict = {}

# Iterate over each username
for username in usernames:
    print(f"Scraping data for {username}...")

    # Get tweets from the user
    tweets = scraper.get_tweets(username, mode='user', number=20)

    # Prepare the list to store tweet data
    tweet_data = []
    for x in tweets['tweets']:
        tweet_info = [username, x['link'], x['text'], x['date'], x['stats']['likes'], x['stats']['comments']]
        tweet_data.append(tweet_info)

    # Store tweet data in dictionary
    tweets_dict[username] = pd.DataFrame(tweet_data, columns=['username', 'twitter_link', 'text', 'date', 'likes', 'comments'])

    # Extract profile information
    profile_info = scraper.get_profile_info(username)

    # Create a DataFrame with profile information
    profile_data = {
        'username': username,
        'image': profile_info['image'],
        'name': profile_info['name'],
        'bio': profile_info['bio'],
        'location': profile_info['location'],
        'website': profile_info['website'],
        'joined': profile_info['joined'],
        'tweets': profile_info['stats']['tweets'],
        'following': profile_info['stats']['following'],
        'followers': profile_info['stats']['followers'],
        'likes_profile': profile_info['stats']['likes']
    }

    profiles_dict[username] = pd.DataFrame([profile_data])

# Create a Pandas Excel writer using XlsxWriter as the engine
with pd.ExcelWriter('twitter_data_multiple_users.xlsx', engine='xlsxwriter') as writer:
    for username in usernames:
        # Write each user's tweets to a separate sheet
        tweets_dict[username].to_excel(writer, sheet_name=f'{username}_tweets', index=False)

        # Write each user's profile information to a separate sheet
        profiles_dict[username].to_excel(writer, sheet_name=f'{username}_profile', index=False)

print("Data saved to 'twitter_data_multiple_users.xlsx'.")
```
